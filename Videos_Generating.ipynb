```python
import cv2
import os
import glob

base_dir = r"D:\Downloads\IP_Project"

frame_dir = os.path.join(
    base_dir,
    "Enhanced_frames"
)

output_video = os.path.join(
    base_dir,
    "Enhanced_frames",
    "Enhanced_video.mp4"
)


# READ ALL ENHANCED FRAMES
frames = sorted(glob.glob(os.path.join(frame_dir, "*.jpg")))

print("Total enhanced frames found:", len(frames))

if len(frames) == 0:
    print("No enhanced frames found.")
    exit()


first_frame = cv2.imread(frames[0])

height, width, layers = first_frame.shape


fps = 30

fourcc = cv2.VideoWriter_fourcc(*'mp4v')

video = cv2.VideoWriter(
    output_video,
    fourcc,
    fps,
    (width, height)
)


for path in frames:

    frame = cv2.imread(path)

    if frame is None:
        print("Skipped:", path)
        continue

    frame = cv2.resize(frame, (width, height))

    video.write(frame)

video.release()

print("===================================")
print("Enhanced video created successfully.")
print("Saved at:")
print(output_video)
print("===================================")
```

```python
import cv2
import numpy as np



video1 = r"D:\Downloads\IP_Project\Road_Damaged_Video.mp4"
video2 = r"D:\Downloads\IP_Project\Enhanced_frames\Enhanced_video.mp4"
video3 = r"D:\Downloads\IP_Project\Segmentation\Segmented_frames\segmentation_video.mp4"

output_video = r"D:\Downloads\IP_Project\Output_video.mp4"


cap1 = cv2.VideoCapture(video1)
cap2 = cv2.VideoCapture(video2)
cap3 = cv2.VideoCapture(video3)


# VIDEO SIZE

width = 640
height = 360

# Final merged size
final_width = width * 3
final_height = height

fps = 20


fourcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter(
    output_video,
    fourcc,
    fps,
    (final_width, final_height)
)


while True:

    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()
    ret3, frame3 = cap3.read()

    if not (ret1 and ret2 and ret3):
        break

    # Resize videos equally
    frame1 = cv2.resize(frame1, (width, height))
    frame2 = cv2.resize(frame2, (width, height))
    frame3 = cv2.resize(frame3, (width, height))

    # Merge horizontally
    merged = np.hstack((frame1, frame2, frame3))

    # Save output
    out.write(merged)

    # Display
    cv2.imshow("Merged 3 Videos", merged)

    # Press Q to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap1.release()
cap2.release()
cap3.release()

out.release()

cv2.destroyAllWindows()

print("DONE")
print("Saved at:", output_video)
```
